# Exercise 1: Proofs of Consistency and Unbiasedness
## (a)
We can rewrite 
$$
\hat{\sigma}_{XY,N} = \frac{1}{N}\sum_{i=1}^N(X_i - \bar{X}_N)(Y_i - \bar{Y}_N) = \frac{1}{N}\sum_{i=1}^N\left(X_iY_i - \bar{X}_NY_i - \bar{Y}_NX_i + \bar{X}_N\bar{Y}_N\right)
$$
$$
= \frac{1}{N}\sum_{i=1}^N Y_iX_i - \bar{X}_N\frac{1}{N}\sum_{i=1}^NY_i - \bar{Y}_N\frac{1}{N}\sum_{i=1}^NX_i + \bar{X}_N\bar{Y}_N
$$
$$
= \frac{1}{N}\sum_{i=1}^N Y_iX_i - \bar{X}_N\bar{Y}_N = \bar{Z}_N - \bar{X}_N\bar{Y}_N,
$$
where we define $Z_i=Y_iX_i$ and $\bar{Z}_N$ is it's sample mean. By WLLN, $\bar{Z}_N\overset{p}{\to}E[Z]=E[YX]$ and $\bar{X}_N\overset{p}{\to}E[X]$ and $\bar{Y}_N\overset{p}{\to}E[Y]$. Consider the function $g(a,b,c) = a - bc$. We see that $g(\bar{Z}_N, \bar{X}_N, \bar{Y}_N)=\hat{\sigma}_{XY,N}$ and that $g(\cdot)$ this is a continuous function. Thus, by the CMT, 
$$
\hat{\sigma}_{XY,N}=g(\bar{Z}_N, \bar{X}_N, \bar{Y}_N)\overset{p}{\to}g(E[Z],E[X],E[Y]) = E[XY] - E[X]E[Y] = Cov[X,Y].
$$

## (b)
$\hat{F}_N(x)$ is a random variable because it is a function of random variables. Define $Z_i=\boldsymbol{1}\{X_i\leq x\}$. Then $\hat{F}_N(x)$ is just the sample mean of $Z_i$. By WLLN, $\hat{F}_N(x)=\bar{Z}_N\overset{p}{\to}E[Z_i]$. Let's take a closer look at this:
$$
E[Z_i] = E[\boldsymbol{1}\{X\leq x\}] = \int_{\infty}^{\infty}\boldsymbol{1}\{a\leq x\}f(x)dx = P(X\leq x) = F(x),
$$
which is just the cdf. We can therefore interpret $\hat{F}_N(x)$ as the empirical cdf. What about unbiasedness?
$$
E[\hat{F}_N(x)] = E\left[\frac{1}{N}\sum_{i=1}^N\boldsymbol{1}\{X_i\leq x\}\right] = \frac{1}{N}\sum_{i=1}^NE[\boldsymbol{1}\{X\leq x\}] = E[\boldsymbol{1}\{X\leq x\}] = P(X\leq x) = F(x),
$$
so that $\hat{F}_N(x)$ is also an unbiased estimator of $F(x)$.

## (c)
Consider first unbiasedness.
$$
E[\hat{\theta}_N] = E[X_1] = E[X],
$$
so that, yes, it is unbiased for $E[X]$. What about consistency? Let's produce a contradiction by looking at a specific case. Let $X\sim U[0,1]$. Fix an $\epsilon > 0$. Then
$$
P(|\hat{\theta}_N - E[X]|>\epsilon) = P(|X_1 - E[X]|>\epsilon) = P(X_1 - E[X] > \epsilon) + P(-X_1 + E[X] > \epsilon) 
$$
$$
= 1 - P(X_1 < \epsilon + E[X]) + P(X_1 < E[X] - \epsilon) = 1 - (\epsilon + E[X]) + E[X] - \epsilon = 1 - 2\epsilon \nrightarrow 0,
$$
which follows since $P(X<x)=x$ when $X\sim U[0,1]$. Thus, $\hat{\theta}_N$ is not consistent for $E[X]$, which should come as no surprise. Since $\hat{\theta}_N = X_1$ it is always the first observation and does not change with the sample size.

## (d)
We have the following expression for the sample mean of $E[Y|X=1]$:
$$
\frac{\sum_{i=1}^NY_iX_i}{\sum_{i=1}^NX_i}.
$$
We can rewrite as follows:
$$
\frac{\frac{1}{N}\sum_{i=1}^NY_iX_i}{\frac{1}{N}\sum_{i=1}^NX_i}.
$$
By applying the WLLN to the numerator and denominator and then CMT, we obtain
$$
\frac{\frac{1}{N}\sum_{i=1}^NY_iX_i}{\frac{1}{N}\sum_{i=1}^NX_i} \overset{p}{\to}\frac{E[YX]}{E[X]}
$$
But since $E[X] = P(X=1)$ and 
$$
E[YX] = E[Y\cdot 1|X=1]P(X=1) + E[Y\cdot 0|X=0] = E[Y|X=1]P(X=1)
$$
(by law of iterated expectations) then
$$
\frac{E[YX]}{E[X]} = \frac{E[Y|X=1]P(X=1)}{P(X=1)} = E[Y|X=1],
$$
and we've finished our proof of consistency.

# Exercise 2:  Constructing Estimators
## (a)
Since $Var[Y|X=x] = E[Y^2|X=x] - E[Y|X=x]^2$, we can suggest the following estimator
$$
\frac{\sum_{i=1}^NY_i^2\boldsymbol{1}\{X_i=x\}}{\sum_{i=1}^N\boldsymbol{1}\{X_i=x\}} - \left(\frac{\sum_{i=1}^NY_i\boldsymbol{1}\{X_i=x\}}{\sum_{i=1}^N\boldsymbol{1}\{X_i=x\}}\right)^2.
$$
For your own amusement, we can prove this estimator is consistent. We know from 1d that the portion in parantheses should converge in probability to $E[Y|X=x]$. The portion on the left, by appling the WLLN to the numerator and the denominator, and then applying the CMT, will converge to
$$
\frac{E[Y^2\boldsymbol{1}\{X=x\}]}{P(X=x)} = \frac{E[Y^2|X=x]P(X=x) + E[Y^2\cdot 0|X\neq x]P(X\neq x)}{P(X=x)} = E[Y^2|X=x].
$$
We can then combine these two results using the CMT to show that our estimator converges in probability to $Var[Y|X=x]$.

## (c)
First we write the likelihood:
$$
f(X_1,X_2,...,X_N) = \prod_{i=1}^Nf(X_i) = \prod_{i=1}^Np^{X_i}(1-p)^{1-X_i}.
$$
Taking logs, we obtain the log likelihood
$$
\sum_{i=1}^N\bigg(X_i\log(p) + (1-X_i)\log(1-p)\bigg).
$$
Taking FOCs w.r.t $p$, we obtain
$$
\sum_{i=1}^N\bigg(\frac{X_i}{\hat{p}_N} - \frac{1-X_i}{1-\hat{p}_N}\bigg) = 0
$$
from which we can rearrange to
$$
\frac{1}{\hat{p}_N}\sum_{i=1}^NX_i = \frac{1}{1-\hat{p}_N}\sum_{i=1}^N(1-X_i),
$$
then multiplying both sides by $(1-\hat{p}_N)\hat{p}_N$, we obtain
$$
\sum_{i=1}^NX_i - \hat{p}_N\sum_{i=1}^NX_i = N\hat{p}_N - \hat{p}_N\sum_{i=1}^NX_i
$$
which implies $\hat{p}_N = \frac{1}{N}\sum_{i=1}^NX_i=\bar{X}_N$. By WLLN, this converges
to $E[X] = p$.

## (b)
First of all, let's ignore the fact that we trimmed the dataset. Why trimming the dataset might be a good thing to do is outside the scope of this class. Just think of it as a good rule of thumb to drop or cap extreme values. Recall from the previous dataset that we only observe wages for those in the work force. Letting $A$ denote whether we observe wages or not, the variance estimates above are consistent estimates for the conditional variance $Var[W_t|A=1]$, where $W_t$ denotes a random variable for wage at year $t$. Our estimate of the sample variance for each year $t$ is

$$
\hat{\sigma}_{W,t}^2 = \frac{1}{N_t-1}\sum_{i=1}^{N_t}(W_{t,i} - \bar{W}_{N,t})^2 = \frac{1}{N_t-1}\sum_{i=1}^{N_t}W_{t,i}^2 - \bar{W}_{N,t}^2
$$

where $(W_{t,i})_{i=1}^{N_t}$ is drawn iid (let's assume this for now) from the distribution $W_t|A=1$. Thus, $\hat{\sigma}_{W,t}^2$ is just a continuous function of two sample means. By WLLN,

$$
\frac{1}{N_t-1}\sum_{i=1}^{N_t}W_{t,i}^2 \overset{p}{\to} E[W_t^2|A=1]
$$

and $\bar{W}_{N,t}\overset{p}{\to}E[W_t|A=1]$. Thus, by the continuous mapping theorem, $\hat{\sigma}_{W,t}^2\overset{p}{\to} E[W_t^2|A=1] - E[W_t|A=1]^2 = Var[W_t|A=1]$. 

## (c)
First consider the latter component: $\sqrt{N}(\bar{W}_N - \mu)(\bar{W}_N - \mu)$. By WLLN, $\bar{W}_N - \mu\overset{p}{\to}0$. By CLT, $\sqrt{N}(\bar{W}_N - \mu)\overset{d}{\to}\mathcal{N}(0,\sigma^2)$. Then, by Slutzky, the whole term converges in probability to 0. Thus, the asymptotics distribution of the first term is the only component we need to worry about.

You should realize that $(W_i - \mu)^2$ is just a random variable with mean $\sigma^2$. Since observations are iid, by CLT, the first component converges in probability to

$$
\mathcal{N}(0, Var[(W_i - \mu)^2]),
$$

and we have our desired result.

Let $\sigma_{1968}^2$ and $\sigma_{2011}^2$ denote the variance in log wages in 1968 and 2011, respectively. Our null hypothesis will be $H_0:\sigma_{1968}^2 - \sigma_{2011}^2 = 0$. Consider the following test statistic:

$$
    \tau_N = \frac{\hat{\sigma}_{N,1968}^2 - \hat{\sigma}_{N,2011} - (\sigma_{1968}^2 - \sigma_{2011}^2)}{\sqrt{\hat{V}_{1968}/N_{1968} + \hat{V}_{2011}/N_{2011}}}.
$$

where

$$
\hat{V}_t = \frac{1}{N_{t}}\sum_{i=1}^{N_{t}}\left[(W_{i,t} - \bar{W}_{N,t})^2 - \frac{1}{N_{t}}\sum_{i=1}^{N_{t}}(W_{i,t} - \bar{W}_{N,t})^2\right]^2
$$

for $t=1968,2011$. If we assume that $\hat{\sigma}_{N,1968}^2$ and $\hat{\sigma}_{N,2011}$ are independent, then $\tau_N\overset{d}{\to}\mathcal{N}(0,1)$, which utilizes our asymptotic analysis in part (c). We can now estimate the p-value under the null hypothesis as

$$
\hat{p} = 2(1-\Phi(|\tau_N|)),
$$

where $\Phi()$ is the standard normal cdf.

```r
for (m in c(0, 1)) {
    W.1968 = df[df$male==m & df$year==1968, "logwage"]
    W.2.1968 = (W.1968 - mean(W.1968, na.rm=TRUE))^2
    V.1968 = var(W.2.1968, na.rm=TRUE)

    W.2011 = df[df$male==m & df$year==2011, "logwage"]
    W.2.2011 = (W.2011 - mean(W.2011, na.rm=TRUE))^2
    V.2011 = var(W.2.2011, na.rm=TRUE)

    s.1968 = var.bygen[var.bygen$male==m & var.bygen$year==1968, "logwage"]
    s.2011 = var.bygen[var.bygen$male==m & var.bygen$year==2011, "logwage"]

    N.1968 = length(W.1968)
    N.2011 = length(W.2011)

    tstat = abs(s.1968 - s.2011)/sqrt(V.1968/N.1968 + V.2011/N.2011)
    
    pval = 2*(1 - pnorm(tstat))
    print(pval)
}
```

We can also compute the p-value using the bootstrap method.

```r
K = 1000 # number of bootstrap draws
set.seed(210)

for (m in c(0, 1)) {
    W.1968 = df[df$male==m & df$year==1968, "logwage"]
    W.2011 = df[df$male==m & df$year==2011, "logwage"]
    
    N.1968 = length(W.1968)
    N.2011 = length(W.2011)
    
    var.1968 = lapply(1:K, function(k) {return(var(sample(W.1968, size=N.1968, replace=TRUE)))})
    var.1968 = unlist(var.1968)
    
    var.2011 = lapply(1:K, function(k) {return(var(sample(W.2011, size=N.2011, replace=TRUE)))})
    var.2011 = unlist(var.2011)
    
    var.m = var.1968 - var.2011
    
    var.null = var.m - mean(var.m)
    
    over = length(var.null[var.null > abs(var(W.1968) - var(W.2011))])
    under = length(var.null[var.null < - abs(var(W.1968) - var(W.2011))])
    
    pval = (over + under)/K
    print(pval)
}
```

Thus, obtain a p-value of 0 using both the bootstrap method and using the analytically constructed test statistics using the asymptotic properties of the sample variance.

## (e)
For a generic random variable $Y$ and another random variable $X$ which contains $J$ discrete values, we can perform the following decomposition of variance.

$$
Var[Y] = E[(Y-E[Y])^2] = \sum_{j=1}^JE[(Y-E[Y])^2|X=j]P(X=j)
$$

$$
 = \sum_{j=1}^JE[Y^2 - 2YE[Y] + E[Y]^2|X=j]P(X=j)
$$

$$
 = \sum_{j=1}^J\bigg(E[Y^2|X=j] + \big(E[Y|X=j]^2 - E[Y|X=j]^2\big) - 2E[Y|X=j]E[Y] + E[Y]^2\bigg)P(X=j)
$$

$$
= \sum_{j=1}^J\bigg(E[Y|X=j]^2 - 2E[Y|X=j]E[Y] + E[Y]^2\bigg)P(X=j) + \sum_{j=1}^J\bigg(E[Y^2|X=j] - E[Y|X=j]^2\bigg)P(X=j)
$$

$$
= \sum_{j=1}^J\bigg(E[Y|X=j] - E[Y]\bigg)^2P(X=j) + \sum_{j=1}^JVar[Y|X=j]P(X=j)
$$

and we have our desired result.

## (f)
Let's consider just the sample of women. The math will carry through for men as well. We can obtain an estimator of the across group variance as

$$
\hat{\sigma}_{N,a}^2 = \sum_{d=0}^1(\bar{W}_{N,d} - \bar{W}_N)^2\frac{N_d}{N}
$$

and the within group variance

$$
\hat{\sigma}_{N,w}^2 = \sum_{d=0}^1\hat{\sigma}_{N,d}^2\frac{N_d}{N},
$$

$N_d$ is the sample size of group $D=d$, $N$ is the total sample size, $\bar{W}_{N,d}$ is the sample mean of log wages for group $D=d$, $\bar{W}_N$ is the sample mean over everyone, and $\sigma_{N,d}^2$ is the sample variance associated with group $d$.

By WLLN $\bar{W}_{N,d}\overset{p}{\to} E[W|D=d]$ and $\bar{W}_N\overset{p}{\to}E[W]$. Also by WLLN, 

$$
\frac{N_d}{N} = \frac{1}{N}\sum_{i=1}^N\mathbb{1}\{D_i=d\} \overset{p}{\to} E[\mathbb{1}\{D=d\}] = P(D=d),
$$

so that by CMT, 

$$
\hat{\sigma}_{N,a}^2\overset{p}{\to} \sum_{d=0}^1(E[W|D=d] - E[W])^2P(D=d).
$$

We can rely on the standard proof of consistency of sample variance to show that $\hat{\sigma}_{N,d}^2\overset{p}{\to}Var[W|D=d]$ so that by CMT,

$$
\hat{\sigma}_{N,w}^2\overset{p}{\to} \sum_{d=0}^1Var[W|D=d]P(D=d).
$$

## (a) Two Sided Test for a Mean
Let $X$ denote a random variable for years of education and $\mu_X$ denote its mean. We have a sample $(X_i)_{i=1}^N$ and therefore can calculate the sample mean of education $\bar{X}_N$. 
Our null hypothesis is $H_0: \mu_X = 12$. Our alternative hypothesis is $H_1:\mu_X\neq 12$. From the class notes, our test statistic is the absolute value of
$$
\tau_N = \frac{\sqrt{N}(\bar{X}_N - 12)}{\hat{\sigma}_N}
$$
where $\hat{\sigma}_N$ is the square root of the sample variance (i.e., the sample standard deviation). The test is $\phi_N = \boldsymbol{1}\{|\tau_N| > z_{1-\alpha/2}\}$. We can compute our t-statistic as follows:

```r
file_url = "https://raw.githubusercontent.com/jtorcasso/teaching/master/econ210_fall2017/data/psid_2011.csv"
df = read.csv(file_url)

c = 12

edu.m = mean(df$edu)
edu.m
```

```r
edu.sd = sd(df$edu)
N = length(df$edu)
t.stat = sqrt(N)*(edu.m - c)/edu.sd
t.stat
```

The p-value is obtained as
$$
\hat{p}_N = 2(1- \Phi(|\tau_N|)),
$$
where $\Phi(\cdot)$ is the standard normal cdf.

```r
p.val = 2*(1 - pnorm(abs(t.stat)))
p.val
```

We can also calculate this p-value using the bootstrap. Heuristically, we build a null distribution for the sample mean and then we calculate the fraction of values in the null distribution more extreme than our estimate of the sample mean.

```r
nsim = 10000
set.seed(210)

b.means = lapply(1:nsim, function(k) {return(mean(sample(df$edu, size=N, replace=TRUE)))})
b.means = unlist(b.means)
b.null = b.means - mean(b.means) + c
outer = length(b.null[abs(c - b.null) > abs(c - edu.m)])
p.val = outer/nsim
p.val
```

And we obtain a p-value very close to the p-value using our test statistic.

## (d) Two-sided test for a Difference in Means
The null hypothesis is $H_0:\mu_X - \mu_Y = 7$ and the alternative hypothesis is $H_1:\mu_X - \mu_Y \neq 7$, where $\mu_X$ and $\mu_Y$ represent the mean wages of men and women. Let $\bar{X}_N$ and $\bar{Y}_N$ denote the sample means of men's and women's wages and $\hat{\sigma}_X$ and $\hat{\sigma}_Y$ their respective sample standard deviations. Also, let $N_X$ and $N_Y$ denote the sample sizes of each males and females in the sample.

Our test statistic will be the absolute value of
$$
\tau_N = \frac{\bar{X}_N - \bar{Y}_N - 7}{\sqrt{\hat{\sigma}_X^2/N_X + \hat{\sigma}_Y^2/N_Y}}.
$$
Our test is $\phi_N = \boldsymbol{1}\{|\tau_N|>z_{1-\alpha/2}\}$. The p-value is $2(1-\Phi(|\tau_N|))$.

```r
file_url = "https://raw.githubusercontent.com/jtorcasso/teaching/master/econ210_fall2017/data/psid_1970.csv"
df = read.csv(file_url)

df$wage =  ifelse(df$inc_labor==0 | df$hours==0, NaN, df$inc_labor/df$hours)
df.trim = df[df$wage > 5 & df$wage < 1000, ]
df.trim = na.omit(df.trim)

# Null difference
c = 7

# Wages of males and females
wages.m = df.trim[df.trim$male==1 & df.trim$edu >= 16, "wage"]
wages.w = df.trim[df.trim$male==0 & df.trim$edu >= 16, "wage"]

# Mean, variance and sample size of male wages
mean.m = mean(wages.m)
var.m = var(wages.m)
N.m = length(wages.m)

# Mean, variance and sample size of female wages
mean.w = mean(wages.w)
var.w = var(wages.w)
N.w = length(wages.w)

mean.m - mean.w
```

```r
# t-statistic
t.stat = (mean.m - mean.w - c)/sqrt(var.m/N.m + var.w/N.w)
t.stat
```

```r
p.val = 2*(1 - pnorm(abs(t.stat)))
p.val
```
