# Random Variables
How do we understand tables of data mathematically? Consider the following table of data collected for 3 rolls of a single 6-faced die.

|i|Roll|
|-|-|
|1|5|
|2|2|
|3|6|
-------

This table represents a realization from a random vector, or a vector of random variables. Conceptually, we can think of our data as just one particular realization of these three dice rolls from a set of possibilities.

Consider just the ith observation. Denote the random variable representing this roll as $X_i$. We can think of a **random variable** as a pairing $(\Omega,P)$, i.e., "a set and a way of describing probabilities $P$ of subsets of $\Omega$." For the dice roll example, the set of possibile outcomes is $\Omega_i=\{1,2,3,4,5,6\}$. $X_i$ is characterized by this set of possibilities and the probabilities of the occurence of every subset of $\Omega_i$. We call these subsets events. For instance, one event could be $A=\{1,2,3\}$ or another $B=\{2\}$. If this is a fair die, then $P(A)=P(X_i\leq 3)=1/2$ and $P(B)=P(X_i=2)=1/6$. 

# Random Vectors
The random vector representing our data table is a vector of random variables $X=(X_1,X_2,X_3)$, which for arbitrarily sized samples can be written as $(X_i)_{i=1}^N$ or just $(X_i)_i$. The set of possible outcomes is now 

$$
\Omega=\Omega_1\times\Omega_2\times\Omega_3 =
$$

$$
 =\{(1,1,1),(2,1,1),...,(6,6,5),(6,6,6)\}
$$

where $\times$ denotes cartesian product. In other words, $\Omega$ is all possible combinations of three dice rolls. Since $X$ is a random vector, there is an associated probability of any event in $\Omega$ occuring. Thus, if each dice roll is independent and the die is fair, $P(X=(1,1,1))=(1/6)^3=1/216$. Another event, denoted by $C\subseteq\Omega$ could be $X_1\in\{1,2,3,4,5,6\}$, $X_2\in\{1,2,3,4,5,6\}$, $X_3=4$, so that $P(C)=1\cdot 1\cdot 1/6=1/6$. 

## Events of Random Variables and Probability
Subsets of the sample space $\Omega$ are called events. Let $X$ be a random variable with sample space $\Omega$. Consider an event $A\subset\Omega$ and another event $B\subset\Omega$. If $\Omega=(-\infty,\infty)$, or the real line, $A$ could be $[1,3]$ and $B$ could be $[2,4]$. Note that everything here follows for random vectors as well. In what follows we will use $P(A)$ as shorthand for $P(X\in A)$. Recall these basic laws of probability:

## Basic Axioms
1. $0\leq P(A)\leq 1$
2. $P(\Omega) = 1$
3. Additivity.
Let $A=\cup_iA_i$, where $(A_i)_i$ are disjoint from one another. Then
$$
P(A) = \sum_iP(A_i)
$$

## Implications of Basic Axioms
1. Let $A^c=\Omega\setminus A$ denote the complement of $A$ (the elements of $\Omega$ not in $A$).
$$
P(A) = 1 - P(A^c)
$$
2. If $A\subseteq B$, then $P(A)\leq P(B)$
3. 
$$
P(A\cup B) = P(A) + P(B) - P(A\cap B)
$$
4. Let $\Omega = \cup_iA_i$, where $(A_i)_i$ are disjoint from one another.
$$
P(B) = \sum_iP(B\cap A_i)
$$

## Definition of conditional probability
$$
P(A|B) = \frac{P(A\cap B)}{P(B)}
$$

which we can use to write Baye's rule

$$
P(A|B) = \frac{P(B|A)P(A)}{P(B)}
$$

## Moving forward
We will typically use the pdf (or pmf, for discrete random variables) to write probabilities. The pdf characterizes the distribution of random variables. We cover this next.

# Distributions of Random Variables
The most common way to characterize random variables is by specifying their distribution. The distribution describes the probability of events. This is similar to our descriptions for the dice roll example above. Two broad classes of distributions are **continuous** and **discrete** distributions (corresponding to continuous and discrete random variables). **Discrete** random variables take on a countable (in this class, just think of finite) number of values. **Continous** random variables take on a continuum of values, such as those over the invertval $[0,1]$ or perhaps the whole real line $(-\infty,\infty)$.

Some common continous distributions are the normal and uniform distributions, denoted $\mathcal{N}(\mu,\sigma^2)$ and $U[a,b]$, respectively. A common discrete distribution is the Bernoulli distribution, denoted $Bernoulli(p)$. The parameters $\mu$, $\sigma^2$, $a$, $b$ and $p$ specify these distributions further. We write $X\sim\mathcal{N}(\mu,\sigma^2)$ to denote that $X$ has a normal distribution with mean $\mu$ and variance $\sigma^2$. The big difference between discrete and continuous distributions is that discrete distributions have mass points, i.e., $P(X=x)$ may be strictly positive. In our dice roll example, $P(X_i=1)=1/6$. But a continuous distribution has no mass points, i.e., $P(X=x)=0$ for all $x$ when $X$ is continuous.

- Note that writing $X=x$ denotes the event "$X$ is realized as the value $x$". We will typically use lower case letters like $x$ denote realizations and $X$ to denote random variables or random vectors.

So how do we typically describe probabilities? You may recall from your statistics courses the probability density function (pdf), the probability mass function (pmf) and the cumulative distribution function (cdf) of a random variable $X$. Let $X=x$ denote the event "$X$ takes on the particular value $x$."

## The cdf
The cdf is written $F(x) = P(X\leq x)$. 

The cdf has the following properties:

- $\lim_{x\to-\infty}F(x)=0$
- $\lim_{x\to\infty}F(x)=1$
- $F$ is continuous from the right
- $F$ is non-decreasing

## The pdf (or pmf) 
We will use $f(x)$ to denote the pmf for discrete random variables and the pdf for continuous random variables.

### Discrete
$$
f(x) = P(X=x)
$$
where $f(x)$ denotes the pdf of $X$ evaluated at $x$.

#### Properties of pmf
1. $f(x)\geq 0$ for all x
2. $\sum_kf(x_k) = 1$ where the sum is over all points over which $X$ takes values
3. We can write probabilities of "$X$ being in a set A" as
$$
P(X\in A) = \sum_k\boldsymbol{1}\{x_k\in A\}f(x_k) = \sum_{x_k\in A}f(x_k)
$$
where $A$ could be a set like $\{1,2\}$ as in our dice roll example. Why does the pmf sum to 1? Because it is equalt to $P(X\in\Omega)=1$.

### Continuous
$$
F(x) = \int_{-\infty}^xf(y)dy
$$
where $f(y)$ denotes the pdf of $X$ evaluated at $y$. When $f(x)$ is continuous, we can simply write
$$
f(x) = \frac{d}{dx}F(x)
$$
which uses the fundamental theorem of calculus. The pdf $f(x)$ captures the likelihood that $X$ takes on the value $x$
or perhaps more intuitively, consider an infinitesmal interval $I_x = [x,x+dx]$. Then we can think of the density as $f(x)\approx P(x\in I_x)/dx$. The intuition of the pdf being proportional to some infinitesmal probability is why many of the properties of probabilities also apply to the density.

#### Properties of pdf
1. $f(x)\geq 0$ for all x
2. $\int_{-\infty}^{\infty}f(x)dx = 1$
3. We can write probabilities of "$X$ being in a set A" as
$$
P(X\in A) = \int_{-\infty}^{\infty}\boldsymbol{1}\{x\in A\}f(x)dx = \int_Af(x)dx
$$
where $A$ could be an interval like $[0,1]$, so that
$$
P(X\in A) = P(X\in[0,1]) = \int_0^1f(x)dx = F(1) - F(0).
$$
(the latter equality really only holds if $A=(0,1]$ because of the right continuity of $F$. But for most cases in this class, like the normal distribution, we can just think of using the closed interval)

Note that if $X$ is continuous, $P(X=x)=0$ for every $x$. That is, $X$ has no "mass points." The intuition behind this fact is that if you were to simulate from a continuous random variable, the likelihood of simulating any given value would be essentially 0.

### Neither discrete and continuous
There are also random variables that are neither discrete nor continuous. Take for example the random variable $Z$, defined as $Z=max\{0,X\}$, where $X\sim U[-1, 1]$. This random variable has a mass point at $Z=0$, i.e., $P(Z=0) = 1/2$, so that it is not continuous. But it also cannot be discrete because it can take values anywhere in the interval $[0,1]$.

These sort of random variables may only come about in this course when we deal with truncated data. In that case, we will look at them in more detail, otherwise, feel safe in assuming we will either be dealing with purely discrete or purely continuous random variables.

## Support
Another important concept is called the "support of a random variable $X$." We define the support as $supp(X) = \{x|f(x) > 0\}$ where $f(x)$ is the pmf when $X$ is discrete and the pdf when $X$ is continuous. It is a useful concept because sometimes continuous random variables don't have support over the whole real line. For instance, consider an exponential distribution, which has pdf

$$
f(x) = 
\begin{cases}
\lambda e^{-\lambda x} & \text{ if } x\geq 0 \\
0 & \text{ if } x < 0.
\end{cases}
= \boldsymbol{1}\{x\geq 0\}\lambda e^{-\lambda x}
$$

for some fixed constant $\lambda > 0$. Thus we can write

$$
1 = \int_{-\infty}^{\infty}\boldsymbol{1}\{x\geq 0\}\lambda e^{-\lambda x}dx = \int_0^{\infty}\lambda e^{-\lambda x}dx
$$

What's the point? Sometimes I get lazy and just want to write the pdf as $f(x) = \lambda e^{-\lambda x}$, but I keep in mind that the support is $[0,\infty)$. One question you might ask, is how is $supp(X)$ different from $\Omega$? I suppose for most continuous random variables we think of $\Omega=(-\infty,\infty)$. But when $X\sim exp(\lambda)$ (an exponential distribution, as above), $supp(X) = [0,\infty)\subseteq\Omega$. So a lot of times our support is a subset of $\Omega$.

## Why use the same notation, f(x), to denote both pdf and pmf?
Why do we use the same notation $f(x)$ to denote the pmf when $X$ is discrete but the pdf when $X$ is continuos? This is confusing, why don't we use separate notation if they are not the same thing!? 

1. Because **we can write probabilities** as integrals of the pdf for continuous random variables and sums (the discrete counterpart to the integral) of the pmf for discrete random variables. 
2. Many rules that hold for pmfs also hold for pdfs, so it is easier to just write use one expression

## Relation between pdf and cdf
Given our analysis above, the cdf relates to the pdf and pmf in the following way:

- Discrete case
$$
P(X\leq x) = \sum_{a\leq x}f(a) = \sum_{a\leq x}P(X=a)
$$

- Continuous case
$$
P(X\leq x) = \int_{-\infty}^xf(a)da
$$

so that we can equivalently write the pdf as $f(x)=\frac{d}{dx}P(X \leq x)=\frac{d}{dx}F(x)$ when $f(x)$ is continuous.

## Examples

### Discrete Distributions

Let $X\sim Bernoulli(p)$. This distribution takes two values, 0 and 1. It takes the value 1 with probability $p$.

- pmf: 
$$
f(x) = 
\begin{cases}
1 - p & \text{ if } x = 0 \\
p & \text{ if } x = 1 \\
0 & \text{ if } x\notin\{0,1\}
\end{cases}
$$

- cdf: 
$$
F(x) = 
\begin{cases}
0 & \text{ if } x < 0 \\
1 - p & \text{ if } x\in[0, 1) \\
1 & \text{ if } x \geq 1
\end{cases}
$$

We can make 10 draws from a Bernoulli(p) random variable in R using the following command.

```r
p = 0.25
rbinom(n=10, size=1, prob=p)
```

### Continuous Distributions

Let $X\sim U[a, b]$. 
- pmf: $f(x) = \frac{1}{b-a}$ for all $x\in[a,b]$, otherwise $f(x)=0$
- cdf: $F(x) = \frac{x-a}{b-a}$ for all $x\in[a,b]$ and $F(x)=1$ for $x>b$ and $F(x)=0$ for $x<a$. Note you can confirm that $f(x) = \frac{d}{dx}F(x) = \frac{1}{b-a}$.

We can make 10 draws from a $U[0,1]$ random variable in R.

```r
runif(10, min=0, max=1)
```

Now let $X\sim\mathcal{N}(\mu,\sigma^2)$, a normal random variable.
- pmf: 
$$
f(x) = \frac{1}{\sqrt{2\pi}\sigma}exp\left(-\frac{1}{2\sigma^2}(x-\mu)^2\right)
$$
- cdf: $F(x) = \int_{\infty}^xf(x)$, no way to write cdf in a closed form.

We can make 10 draws from a normal random variable with $\mu=1$ and $\sigma=2$ in R using the following command.

```r
rnorm(10, mean=1, sd=2)
```

## Why do we care about distributions?
So what's the point? A sample of data from some distribution will help us learn about this unknown distribution. For example, $(X_1,...,X_N)$ is a sample from the same distribution as $X\sim Bernoulli(p)$, we can learn about $p$. Or if $(X_1,...,X_N)$ is a sample from the $X\sim\mathcal{N}(\mu,\sigma^2)$, we can learn about $\mu$ and $\sigma^2$. Knowing $\mu$, $\sigma^2$ or $p$ (or still other characteristics of these distributions) could be informative for economic parameters we are interested in.

## Comparing true distributions to empirical counterparts
To make these ideas concrete, let us plot a histogram of random draws from a standard normal distribution. We will compare the histogram to the true probability density function. A sample can teach us a lot about the shape of the true underlying distribution. This is especially true for large samples. We will often refer to population objects (like distributions, means and variances) and their sample or empirical counterparts (like a histogram of an empirical distribution, sample means and sample variances). We call them counterparts because the sample objects will in some way "converge" to their population counterparts. We will formalize this notion later, but now we can think that if we had "infinite" data, i.e, a sample of infinite size, we could know the population parameters exactly.

Now let's make draws from a standard normal distribution. If we plot the histogram of these draws (i.e., the frequency of draws), we obtain something that looks like the pdf of standard normal distribution. This exercise shows how data can be thought of as draws from an underlying distribution. And that as we take more draws, we can better approximate the population pdf using a histogram.

```r
set.seed(210)
N = 100

X1 = rnorm(N, mean=0, sd=1)
hist(X1, freq=FALSE, breaks=10,
     main="PDF of Standard Normal and Empirical PDF, N=100")
lines(sort(X1), y=dnorm(sort(X1), mean=0, sd=1))
```

We can also plot the standard normal cdf against its empirical counterpart, the **empirical cdf**, defined as

$$
\hat{F}(x) = \frac{1}{N}\sum_i\boldsymbol{1}\{X_i\leq x\}.
$$

where $\boldsymbol{1}\{\cdot\}$ is an indicator function taking a value of 1 if the argument evaluates to True and 0 if False.

```r
plot(sort(X1), y=pnorm(sort(X1), mean=0, sd=1), type='l',
    main="CDF of Standard Normal and Empirical CDF, N=100",
    xlab="X", ylab="P(X<x)")
lines(seq(-3, 3, length.out=100), ecdf(X1)(seq(-3, 3, length.out=100)))
```

Now see what happens when we make $N=10000$.

```r
set.seed(210)
N = 10000

X1 = rnorm(N, mean=0, sd=1)
hist(X1, freq=FALSE, breaks=40,
     main="PDF of Standard Normal and Empirical PDF, N=10000")
lines(sort(X1), y=dnorm(sort(X1), mean=0, sd=1))
```

```r
plot(sort(X1), y=pnorm(sort(X1), mean=0, sd=1), type='l',
    main="CDF of Standard Normal and Empirical CDF, N=10000",
    xlab="X", ylab="P(X<x)")
lines(seq(-3, 3, length.out=100), ecdf(X1)(seq(-3, 3, length.out=100)))
```

When the sample size grows, we learn more and more information about the underlying distribution from which the sample was drawn. We can see this above, as the histograms and empirical cdfs plotted from a larger sample more closely resemble the standard normal pdf and cdf. 

# (Joint) Distributions of Random Vectors

A random vector also has a distribution, which we call a joint distribution. We will consider the bivariate case for a random vector $(X,Y)$. We may sometimes refer to the case with just a single random variable as the univariate case.

## cdf
The cdf is defined as $F(x,y) = P(X\leq x, Y\leq y)$ with similar properties to the univariate case.

## pdf (of pmf)

We can define the pmf and pdf for a bivariate (joint) distribution $(X,Y)$.

- Discrete case
$$
f(x,y) = P(X=x,Y=y)
$$
where we use $f(x,y)$ to denote the pmf.

- Continuous case
$$
F(a,b) = \int_{-\infty}^a\int_{-\infty}^bf(x,y)dxdy
$$
where we use $f(x,y)$ to denote the pdf. If $f(x,y)$ is continuous, then we can define $f(x,y)$ simply as

$$
f(x,y) = \frac{\partial^2}{\partial x\partial y}F(x,y)
$$

Knowing the pdf and pmf allow us to write probabilities, like in the univariate case.

- Discrete
$$
P((X,Y)\in \{0,1\}\times\{0,1\}) = f(0,0) + f(1,0) + f(0,1) + f(1,1)
$$
where $f(x,y)$ denotes the joint pmf.

- Continuous
$$
P((X,Y)\in [0,1]\times[0,1]) = \int_0^1\int_0^1f(x,y)dxdy
$$
where again $f(x,y)$ denotes the joint pdf. The intuition of this probability is that of taking the volume under the joint density pdf in the block defined by the region from 0 to 1 on the x-axis and from 0 to 1 on the y-axis.

## Examples

### Discrete Distributions
Let $(X,Y)$ denote the outcome of two dice rolls and let each dice roll be independent of the other. Then the pmf is $f(x,y) = (1/6)^2 = 1/36$ for all $(x,y)$ such that $x,y\in\{1,2,3,4,5,6\}$. The cdf is

$$
F(x,y) = \sum_{a\leq x, b\leq y}f(a,b)
$$

so that $F(1, 2) = f(1, 1) + f(1, 2) = 1/36 + 1/36 = 2/36 = 1/18$.

### Continuous Distributions
Let $(X,Y)$ be a bivariate normal with a $2\times 1$ vector of means $\mu = (\mu_X, \mu_Y)'$ and a $2\times 2$ variance matrix

$$
\Sigma = 
\begin{pmatrix}
\sigma_X^2 & \sigma_{XY} \\
\sigma_{XY} & \sigma_Y^2
\end{pmatrix}.
$$

Then we can write the joint pdf as
$$
f(x,y) = \frac{1}{\sqrt{|(2\pi)^2\Sigma|}}exp\left(-\frac{1}{2}\big((x,y)'-\mu\big)'\Sigma^{-1}\big((x,y)'-\mu\big)\right),
$$
where for a generic matrix $W$, $|W|$ denotes the determinant of $W$ and $W'$ the transpose of $W$. The cdf cannot be written down in a closed form.

We can draw from a bivariate normal distribution using the following commands.

```r
library(MASS) # for simulating from multivariate normal

mvrnorm(10, mu=c(1, 1), Sigma=cbind(c(1, 0), c(0, 1)))
```

## Comparing the true pdf to its population counterpart (bivariate case)
Now let's visualize the bivariate pdf. We can perform a similar exercise as we did for the univariate case by repeatedly drawing from a random vector $(X, Y)$. This way we can produce an empirical joint pdf (histogram). Below we plot the empirical and population pdfs of a bivariate normal distribution with zero mean, variance 1, and covariance equal 0.5.

```r
library(mvtnorm) # for multivariate normal distribution
library(plot3D) # for 3d plotting

set.seed(210)
N = 1000
G = 50
B = 20

mu = c(0, 0)
sigma = cbind(c(1, 0.5), c(0.5, 1))

par(mfrow=c(1, 2))
# 3d histogram of draws from bivariate normal pdf
draws = mvrnorm(N, mu=mu, Sigma=sigma)
x_c = cut(draws[,1], B)
y_c = cut(draws[,2], B)
z = table(x_c, y_c)
hist3D(z=z, border="black", theta=30, phi=30,
      main="Bivariate histogram")

# 3d visual of bivariate normal pdf
x = seq(-3, 3, by=6/G)
y = seq(-3, 3, by=6/G)
f = function(x, y) {dmvnorm(cbind(x, y), mean=mu, sigma=sigma)}
z = outer(x, y, f)
persp3D(x, y, z, theta=30, phi=30, facets=FALSE, space=0.9, border=NA,
     main="Bivariate normal pdf")
```

# A Note on Notation
Once we start dealing with joint distributions, as in the bivariate case with $(X,Y)$, we need to be able to differentiate between the pdf or cdf of each random variable separately, and the joint pdf or cdf. Thus, let $f_X(x)$ denote the pdf of $X$, sometimes referred to as the **marginal** distribution of $X$. Similarly, denote $F_X(x)$ as the marginal cdf. Let $f_{X,Y}(x,y)$ denote the joint pdf and $F_{X,Y}(x,y)$ the joint cdf.

Also, on future occasions I may start to use $X=(X_1,...,X_K)'$ to denote a random vector of length $K$. Then writing joint distributions will be easier, using $f_X(x)$ instead of $f_{X_1,...,X_K}(x_1,...,x_K)$. Hopefully the context should make it clear, but if not don't hesitate to ask!

# Relationship between Marginal and Joint Distribution
We can obtain the marginal distribution by integration or summation (discrete case).

- Discrete
$$
f_X(x) = \sum_yf_{X,Y}(x,y)
$$
where $f_{X,Y}(x,y)$ denotes the joint pmf. **Intuition**: $f_x(x) = P(X=x)$, but we can think of this as really being $P(X=x\text{ and Y is anything})$ and we know from probability that the probability of the union of disjoint subsets equals the sum of the probabilities.

- Continuous
$$
f_X(x) = \int_{-\infty}^{\infty}f_{X,Y}(x,y)
$$
where $f_{X,Y}(x,y)$ denotes the joint pdf.

# Conditional Distributions
What is the distribution of $Y$ conditional on $X$ taking on a certain value $x$? The slice of the joint distribution (see below) of $Y$ and $X$ at $X=x$ will tell us about this. Mathematically, we define the conditional pdf (or pmf, if discrete) as

$$
    f_{Y|X}(x,y) = \frac{f_{X,Y}(x,y)}{f_X(x)}.
$$

An easy way to remember this definition is that you know densities need to integrate (or sum, if discrete) to 1. So suppose we fix $X=x$ to be a particular value. When we integrate the conditional pdf $f_{Y|X}(x,y)$ w.r.t $y$, we should get 1. In order for a density of Y for a fixed $X=x$ to integrate to 1, we need to divide by $f_X(x)$. To see yourself, notice

$$
\int_{-\infty}^{\infty}\frac{f_{X,Y}(x,y)}{f_X(x)}dy = \frac{1}{f_X(x)}\int_{-\infty}^{\infty}f_{X,Y}(x,y)dy = \frac{1}{f_X(x)}f_X(x) = 1.
$$

The discrete case follows similarly. Below we plot the pdf of $Y$ conditional on $X=1$ against the joint distribution of $X$ and $Y$ for a bivariate normal distribution.

```r
library(scatterplot3d)

par(mfrow=c(1, 2))

# 2d pdf in 3d axes
y = seq(-3, 3, length.out=1000)
x = rep(1, 1000)
z = dnorm(1)*dnorm(y, mean=0.5, sd=sqrt(0.75))
scatterplot3d(x=x, y=y, z=z, type="l", ylim=c(-3, 3), xlim=c(-3, 3),
             asp=45, angle=400)

# 3d visual of bivariate normal pdf
x = seq(-3, 3, by=6/G)
y = seq(-3, 3, by=6/G)
f = function(x, y) {dmvnorm(cbind(x, y), mean=mu, sigma=sigma)}
z = outer(x, y, f)
persp3D(x, y, z, theta=30, phi=30, facets=FALSE, space=0.9, border=NA,
     main="Bivariate normal pdf")
```

On the left we take a slice out of the joint distribution of $(X,Y)$ where $X=1$. The left plot is (almost) the conditional distribution of $Y$ given $X=1$, i.e., $f_{Y|X}(y,1)$, except we have not divided by $f_{X}(1)$. 

Note that we will also have sample counterparts to these population conditional distributions. But it can get a bit tricky to plot them for continuous distributions. Why? If we were to try to plot an empirical counterpart to the conditional pdf above, it is likely we have no data in the sample where $X$ is exactly 1 ($X=1$) when $X$ is a continuous random variable. So what we normally do is consider conditioning on a "bin" or "cell" (like an interval) of $X$ values. 

# Independence
Now that we have introduced the concept of joint and conditional distributions we can talk about the dependence between random variables. We say that a $X$ and $Y$ are independent if

$$
f_{X,Y}(x,y) = f_X(x)f_Y(y).
$$

Is $F_{X,Y}(x,y) = F_X(x)F_Y(y)$ if $X$ and $Y$ are independent?

From what we know about conditional distributions this implies that

$$
f_{Y|X}(x,y) \equiv \frac{f_{Y,X}(x,y)}{f_X(x)} = \frac{f_Y(y)f_X(x)}{f_X(x)} = f_Y(y),
$$

that is, the conditional distribution of $Y$ does not depend on the realization of $X$. There are other notions of independence which are implied by (but do not imply) independence. For instance, mean independence $E[Y|X=x]=E[Y]$ and orthogonality or zero covariance $Cov[X,Y] = 0$.

- Also important to note that independence is not a transitive property, i.e., if $X$ is independent of $Y$ and $Y$ is independent of $Z$, then $X$ and $Z$ MAY STILL be DEPENDENT

Notice that it can be difficult to "see" independence by looking at the joint distribution. Below we plot the conditional pdf of $Y$ for several values of $x$.

```r
x.vals = c(-1, 0, 1)
cols = c("red", "blue", "green")

for (i in 1:3) {
    y = seq(-3, 3, length.out=1000)
    x = rep(x.vals[i], 1000)
    z = dnorm(y, mean=0.5*x.vals[i], sd=sqrt(0.75))
    
    if (i == 1){
        s3d <- scatterplot3d(x=x, y=y, z=z, type="l", ylim=c(-3, 3), xlim=c(-3, 3),
                 angle=-80, color=cols[i])
    }
    else {
        s3d$points3d(x=x, y=y, z=z, type="l", ylim=c(-3, 3), xlim=c(-3, 3),
                 col=cols[i])
    }
     
}
```

Above, we see that for the joint pdfs its hard to spot the dependence between $X$ and $Y$, but when we look at the conditional pdfs for different values of $x$, we clearly see that the conditional pdfs of $y$ are shifting as $x$ changes, thus showing that $X$ and $Y$ are dependent. When there is independence, the conditional pdfs of $Y$ would be exactly the same for difference values of $x$.

#  A Note on Joint Densities for Mixed Discrete and Continuous RV
Consider the joint distribution for a random vector $(X,Y)$. What if $Y$ is continuous and $X$ is discrete? In this case, its not really "proper" to write the density as $f(x,y)$, but we will do it anyway. When we write probabilities we will write in the slightly abused fashion

$$
P(X\in A, Y\in B) = \sum_A\int_Bf_{X,Y}(x,y)dy.
$$

For instance, one common situation is $X\in\{0,1\}$ and $Y\in(-\infty, \infty)$. We can write

$$
P(X=1, Y\in[0,1]) = \int_0^1 f_{X,Y}(1,y)dy.
$$

Still more common is to consider conditional probabilities.

$$
f_{Y|X}(x,y) = \frac{f_{X,Y}(x,y)}{f_X(x)} = \frac{f_{X,Y}(x,y)}{P(X=x)}
$$

Just to reiterate, writing a density as $f_{X,Y}(x,y)$ is an abuse of notation because the density doesn't actually exist (since $X$ is discrete), but one that we will typically make. Later in the course we'll see why it doesn't matter too much and that its just useful shorthand to do this.

# Truncation
What about truncated distributions? For example, consider a random variable $X$. What is the distribution of "$X$ given $X > c$?" If we want to write the density of $X|X>c$, we can just remember that what we get must integrate to 1, and for all $x>c$, $f_X(x)$ should be proportional to $f_{X|X>c}(x)$, so that we see that we must have

$$
f_{X|X>c}(x) = \frac{f_X(x)\boldsymbol{1}\{X>c\}}{P(X > c)} = \frac{f_X(x)\boldsymbol{1}\{X>c\}}{1 - P(X \leq c)} = \frac{f_X(x)\boldsymbol{1}\{X>c\}}{1 - F_X(c)}.
$$

You can confirm that this truncated density will integrate to 1 over $(-\infty, \infty)$.
